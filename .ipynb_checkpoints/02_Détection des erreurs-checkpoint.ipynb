# Etude des trajectoires

```python
import matplotlib.pyplot as plt
import pandas as pd
import re
import numpy as np
from mpl_toolkits.mplot3d import axes3d
```

```python
def show_2D(df, msg=""):
    plt.title("id_traj : %d %s" % (df.iloc[0]['id'], msg))
    plt.plot(df['X'], df['Y'], 'r')

    min_x = min(df['X'])
    max_x = max(df['X'])

    min_y = min(df['Y'])
    max_y = max(df['Y'])
    plt.axis([min_x - 20, max_x + 20, min_y - 20, max_y + 20])
    plt.plot(df['X'], df['Y'], 'b+')
    plt.tight_layout()
    return plt
```

```python
file = open("Data/Example.txt", "r")
file.readline() # on ignore la première ligne
time, X, Y, Z, id_traj = [], [], [], [], []

for line in file:
    data = re.split('\s', line)
    id_traj.append(int(data[0]))
    time.append(data[1])
    X.append(float(data[2]))
    Y.append(float(data[3]))
    Z.append(float(data[4]))

df = pd.DataFrame(
    {
        "id": id_traj,
        "X": X,
        "time": time,
        "Y": Y,
        "Z": Z
    }
)
```

### Hypothèses
L'idée ici est de comparer l'angle formé par deux vecteurs de points.
Pour cela nous allons utiliser le Produit Scalaire.

#### Petit rappel sur les produits scalaires
Soit deux vecteurs `A` et `b`, le produit scalaire de ceux deux vecteurs est :

$
\overrightarrow{A} \cdot \overrightarrow{B} = 
||\overrightarrow{A}|| * ||\overrightarrow{B}|| *
    \cos(
        \overrightarrow{A} , \overrightarrow{B}
    )
$

Soit
         
$
\cos(\overrightarrow{A},\overrightarrow{B}) = 
\dfrac
    {\overrightarrow{A} \cdot \overrightarrow{B}}
    {||\overrightarrow{A}|| \cdot ||\overrightarrow{B}||}
$

Soit

$
\angle\overrightarrow{A},\overrightarrow{B} =
\arccos
    \Bigg(
        \dfrac
            {\overrightarrow{A}\cdot \overrightarrow{B}}
            {||\overrightarrow{A}|| \cdot ||\overrightarrow{B}||}
    \Bigg)
$

Voici la fonction permettant de calculer l'angle, dont les valeurs de x et y des vecteurs sont données en paramètres, renvoyant la valeur absolue de l'angle (pour rendre les choses plus simples).
Réalisons des tests simples.

```python
import math
def get_angle_vectors(first_vector, second_vector, _3D = False, id_set=False, _id=28):
    cross_product = first_vector[0] * second_vector[0] + first_vector[1] * second_vector[1]
    
    frst_vec_magnitude_squr = (first_vector[0] ** 2 + first_vector[1] ** 2)
    scnd_vec_magnitude_squr = (second_vector[0] ** 2 + second_vector[1] ** 2)
    
    if _3D:
        cross_product +=  first_vector[2] * second_vector[2]
        frst_vec_magnitude_squr += first_vector[2] ** 2
        scnd_vec_magnitude_squr += second_vector[2] ** 2
        
    magnitude = math.sqrt(frst_vec_magnitude_squr * scnd_vec_magnitude_squr)
    
    try:
        angle = math.degrees(math.acos(cross_product / magnitude))
    except ValueError:
        angle = 0

    if angle < 0.01:
        angle = 0
        
    return angle
```

```python
print(get_angle_vectors([1, 1], [1, 1]))
print(get_angle_vectors([1, 0], [0, 1]))
print(get_angle_vectors([0, 1], [1, 0]))
print(get_angle_vectors([-1, 1], [1, 0]))
print(get_angle_vectors([-1, 0], [1, 0]))
```

### Test sur le jeu de données
Nous réaliserons nos tests sur les 10 premières trajectoires du jeu de données pour commencer, groupant celles qui sont erronées ensemble et le reste ensemble.

```python
def get_vectors_from_df(df, index, _3D=False):
    size = 3 if _3D else 2
    first_vector = [0 for i in range(size)]
    second_vector = [0 for i in range(size)]
    
    x_previous = df['X'].iloc[index - 1]
    x = df['X'].iloc[index]
    x_next = df['X'].iloc[index + 1]

    y_previous = df['Y'].iloc[index - 1]
    y = df['Y'].iloc[index]
    y_next = df['Y'].iloc[index + 1]

    first_vector[0] = x_previous - x
    first_vector[1] = y_previous - y
    second_vector[0] = x_next - x
    second_vector[1] = y_next - y

    if _3D:
        z_previous = df['Z'].iloc[index - 1]
        z = df['Z'].iloc[index]
        z_next = df['Z'].iloc[index + 1]

        first_vector[2] = z_previous - z
        second_vector[2] = z_next - z
            
    return first_vector, second_vector
            

def get_angles(df, _3D=False):
    ar_angles = []
    for j in range(1, len(df) - 1):
        first_vector, second_vector = get_vectors_from_df(df, j, _3D)
        angle = abs(get_angle_vectors(first_vector, second_vector, _3D))
        ar_angles.append(angle)
        
    return ar_angles
```

```python
id_traj_test = id_traj_bad + id_traj_good

def show_traj_health(array_traj):
    plt.figure(figsize=(13, 5))
    for i in array_traj:
        count_error = 0
        df_test = df[df['id'] == i]
        array_angle = get_angles(df_test)
        for angle in array_angle:
            if angle < 50:
                count_error += 1

        plt.subplot(2, 6, array_traj.index(i) + 1)
        if count_error > 0:
            show_2D(df_test, "(erreur)")
        else:
            show_2D(df_test, "(ok)")
            
show_traj_health(id_traj_test)
```

## Amélioration de la détection
Ces groupements de points peuvent représenter par exemple l'atterrissage de l'avion, pour qu'il aille à son hangar. Les nombreux détourset voies qu'un avion peut prendre peut expliquer potentiellement expliquer ses points.  
Tout en restant sur cette idée d'angle, nous allons améliorer notre détection en prenant en compte les valeurs moyennes des angles et de vitesse angulaire, en fonction de l'altitude. Cette dernière sera ainsi subdiviser en trois groupes d'altitudes : de 0 à 100 (altitude supposé proche de l'atterrissage), de 100 à 250 (altitude moyenne), et enfin de 250 à 400.

```python
def get_angles_group_by_altitude(df, _id=28, id_set=False, _3D=False):
    array_angles = [[] for i in range(3)]
    
    for j in range(1, len(df) - 1):
        
        first_vector, second_vector = get_vectors_from_df(df, j)

        angle = abs(get_angle_vectors(first_vector, second_vector, id_set, _id, _3D))
        
        if df['Z'].iloc[j] < 100:
            array_angles[0].append(angle)
        elif df['Z'].iloc[j] < 250:
            array_angles[1].append(angle)
        else:
            array_angles[2].append(angle)
            
        if math.isnan(angle):
            array_angles.append(0)
            
    return array_angles
```

```python
# _ids = [29]
_ids = [i for i in range(50)]
#_ids = [i for i in range(1000)]
total_array_angle = [[] for i in range(3)]
for _id in _ids:
    df_id = df[df['id'] == _id]
    array_angles = get_angles_group_by_altitude(df_id, _id=_id, _3D=True)
    total_array_angle[0] += array_angles[0]
    total_array_angle[1] += array_angles[1]
    total_array_angle[2] += array_angles[2]

avg = [0 for i in range(3)]

# pour éviter le warning si l'array est vide
avg[0] = 0 if len(total_array_angle[0]) == 0 else np.mean(total_array_angle[0])
avg[1] = 0 if len(total_array_angle[1]) == 0 else np.mean(total_array_angle[1])
avg[2] = 0 if len(total_array_angle[2]) == 0 else np.mean(total_array_angle[2])

def to_bars(x, y, y_legende, title='Moyenne par altitude'):
    objects = (x)
    y_pos = np.arange(len(objects))
    performance = y
    plt.bar(y_pos, performance, align='center', alpha=0.5, width=0.4)
    plt.xticks(y_pos, objects)
    plt.ylabel(y_legende)
    plt.title(title)

    plt.show()
    
_altitudes = ('0-100', '100-250', '250-400')
to_bars(_altitudes, [avg[0], avg[1], avg[2]], 'Moyenne vitesse angulaire')
```

```python
def hms_to_seconds(t):
    h, m, s = [float(i) for i in t.split(':')]
    return 3600*h + 60*m + s


def angular_speed(df):
    if df.empty:
        return 0
    array_angle = get_angles(df, _3D=True)
    start_time = df['time'].iloc[0]
    end_time = df['time'].iloc[-1]

    s_start_time = hms_to_seconds(start_time)
    s_end_time = hms_to_seconds(end_time)
    s_delta_time = s_end_time - s_start_time
    
    avg_angle_speed = np.sum(array_angle) / s_delta_time
    return avg_angle_speed

def angular_speed_by_altitude(df):
    ang_speed_by_alt = [0 for i in range(3)]
    if df.empty:
        return ang_speed_by_alt
    ang_speed_by_alt = [0 for i in range(3)]
    array_angle = get_angles(df, _3D=True)
    start_time = df['time'].iloc[0]
    end_time = df['time'].iloc[-1]
    
    ang_speed_by_alt[0] = angular_speed(df[(df['Z'] < 100)])
    ang_speed_by_alt[1] = angular_speed(df[(df['Z'] > 100) & (df['Z'] < 250)])
    ang_speed_by_alt[2] = angular_speed(df[(df['Z'] > 250) & (df['Z'] < 400)])
    
    return ang_speed_by_alt
    
avg = [0 for i in range(3)]
for _id in [i for i in range(28, 50)]:
    if df.empty:
        continue
    ang_speed_by_alt = angular_speed_by_altitude(df[df['id'] == _id])
    avg = np.add(avg, ang_speed_by_alt)


to_bars(_altitudes, avg, 'angular speed avg')
```
